```python
import numpy as np, pandas as pd
import miceforest as mf
```

## Reading Dataset

```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/LLC Assignment/data_preparation.csv')
data
```

```python
data.describe()
```

```python
data.dtypes
```

## Data Preprocessing

```python
data.Year.unique()
```

```python
# drop all rows that contain 'nan' in year column
id = data[data['Year'].isnull()].index[0]
data = data.drop(data.index[id])

data.Year.unique()
```

```python

data['households'] = data['households'].replace('.',np.nan)
data['households'] = data['households'].fillna(value=-1)
data['households'] = data['households'].astype(float)
data['households'] = data['households'].replace(-1,np.nan)


data = data.drop(['DATE'], axis=1)
data.columns
```

## Missing Value Imputation

```python
## Missing Value Imputation:

imputed = mf.ImputationKernel(
  data=data,
  datasets=1,
  save_all_iterations=True,
  random_state=0
)
# Run the MICE algorithm for 5 iterations on each of the datasets
imputed.mice(n_estimators=80, iterations=15,verbose=True)

mice_imputation = imputed.complete_data()
```

```python
## Checking the effect of imputation on columns with missing values:

nan_columns = data.columns[data.isnull().any()].tolist()
for val in nan_columns:
  print("Column {}:".format(val))
  print("RMSE without imputation: {}".format(np.std(data[val])))
  print("RMSE with imputation: {}".format(np.std(mice_imputation[val])))
  print("--------------------------------------------")
  print("--------------------------------------------")
```

```python
# Replacing missing with imputed values:
data = mice_imputation
data.dtypes
```

```python
data
```

```python
# Convert columns to nearest integer and convert into integer data type for integer typed columns:

data['Year'] = data['Year'].round().astype('int64')
data['Month'] = data['Month'].round().astype('int64')
data['households'] = data['households'].round().astype('int64')
data['gdp'] = data['gdp'].round().astype('int64')
data['new_const_units'] = data['new_const_units'].round().astype('int64')
data.dtypes
```

```python
# Saving cleaned preprocessed data in the form of csv:
data.to_csv('/content/drive/MyDrive/Colab Notebooks/LLC Assignment/data_preprocessed.csv', index=False)
```
